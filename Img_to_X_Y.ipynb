```python
import numpy as np
import pandas as pd
import cv2
import os
import pickle
```

```python
images = []
folder = 'train'
i = 0
files = os.listdir(folder)
files.sort()
for filename in files :
    img = cv2.imread(os.path.join(folder,filename))
    if img is not None:
        images.append(img)
    print(str(i) + '/' + str(len(files)))
    i = i+1
```

```python
pickle_out = open('X.pkl', 'wb')
pickle.dump(images, pickle_out)
pickle_out.close()
```

```python
labels = pd.read_csv('labels.csv').to_numpy()
Y = labels[:,1]
Id = labels[:,0]
```

```python
pickle_out = open('Y_and_Id.pkl', 'wb')
pickle.dump((Y, Id), pickle_out)
pickle_out.close()
```
